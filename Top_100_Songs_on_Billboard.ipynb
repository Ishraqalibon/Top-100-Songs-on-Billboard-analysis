```python
singer_ranking={}
for index, row in data.iterrows():
  singer=row['Singer']
  weeks_on_chart=row['Weeks on Chart']
  if singer in singer_ranking:
    singer_ranking[singer]+=weeks_on_chart
  else:
    singer_ranking[singer]=weeks_on_chart
```

```python
singer_ranking
```

```python
weeks_on_char_singer=pd.DataFrame(list(singer_ranking.items()),columns=['Singer','Weeks on Chart']).sort_values('Weeks on Chart',ascending=False)
weeks_on_char_singer.head(5)
```

```python
plt.figure(figsize=(14,5))
ax=sns.barplot(x=weeks_on_char_singer['Singer'].head(5),y=weeks_on_char_singer['Weeks on Chart'].head(5),color=sns.color_palette()[4])
plt.title('Top 5 Singer with Most Weeks on Chart')
plt.xlabel('Singer')
plt.ylabel('Weeks on Chart')
plt.show
```
